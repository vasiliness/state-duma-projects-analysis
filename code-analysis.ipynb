# Анализ социально значимых законопроектов Государственной Думы VII созыва

## Паспорт проекта
Количественные и качественные данные законопроектов Госдумы VII созыва.

Данные собраны с сайтов: 
1. http://duma.gov.ru/legislative/lawmaking/socznach/ – сайт Госдумы;
2. https://sozd.duma.gov.ru/ – система обеспеченности законодательной деятельности


Описание данных:
- Номер законопроекта – регистрационный номер, который присваивается законопроекту и указывается вместе с наименованием в течение всего периода прохождения законопроекта в Государственной Думе. Важен для работы базы данных.
- Категория – деление законопроектов на категории. В данном датасете их 5.
- Тема – распределение законропроектов на темы. В датасете их 20.
- Название на сайте ГД – уникальные названия с сайта Гос.Думы.
- Название – официальное название закона. 
- Дата принятия – день принятия законопроекта Государственной Думой в окончательной редакции.
- Время рассмотрения – количество дней с внесения до принятия. 
- Субъект права законодательной инициативы	– субъекты, на федеральном уровне определенные в ст. 104 Конституции РФ. Ими являются Президент РФ, СФ, члены СФ, депутаты ГД, Правительство РФ, законодательные (представительные) органы субъектов РФ; Это право специально уполномоченных государственных органов и должностных лиц внести проект закона в парламент (палату парламента) государства. Этому праву соответствует обязанность парламента рассмотреть законопроект и вынести по нему решение.
- Форма законопроекта – основные виды законопроектов: федеральный закон или конституционные федеральные законы. 
- Ответственный комитет – уведомляет субъект права законодательной инициативы, внесший законопроект, либо его представителя, субъекты права законодательной инициативы, внесшие поправки к рассматриваемому законопроекту, либо их представителей.
- Комитеты-соисполнители – комитеты, предоставляяющие свои отзывы на работу.
- Отрасль законодательства – часть системы законодательства, предназначенная для регулирования определенных сфер деятельности. 
- Тематический блок законопроектов – разделение законопроектов на тематические блоки (например, социальная политика).
- Профильный комитет – По итогам первого чтения проект передается в профильный комитет, который готовит его ко второму чтению. Также устанавливаются сроки подачи поправок и дополнений к исходному тексту.
- Предмет ведения – сфера общественных отношений, регулирование которой отнесено Конституцией Российской Федерации исключительно к компетенции Российской Федерации. Под «совместным ведением» понимается отнесение определенных вопросов к компетенции как Федерации, так и ее субъектов.
- Описание – описание законопроекта на сайте Госдумы.
- Польза гражданам – слова о пользе для граждан из-за законопроекта на сайте Госдумы.
- Польза государству – слова о пользе для государства из-за законопроекта на сайте Госдумы.
